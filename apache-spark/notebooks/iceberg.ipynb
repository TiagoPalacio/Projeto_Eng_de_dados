# Apache Iceberg com Apache Spark — Cenário E-commerce

Mesma fonte de dados do Delta Lake, implementada com Apache Iceberg.

## Diferenciais do Iceberg
- Time travel por versão e por timestamp
- Schema evolution sem reescrever dados
- Partition evolution
- Open format (funciona com Spark, Flink, Trino, etc.)

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
from datetime import date

spark = SparkSession.builder \
    .appName("Iceberg - E-commerce") \
    .config("spark.sql.extensions",
            "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions") \
    .config("spark.sql.catalog.local",
            "org.apache.iceberg.spark.SparkCatalog") \
    .config("spark.sql.catalog.local.type", "hadoop") \
    .config("spark.sql.catalog.local.warehouse", "/tmp/iceberg/warehouse") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")
print("SparkSession com Iceberg iniciada!")

# Criar database
spark.sql("CREATE DATABASE IF NOT EXISTS local.ecommerce")
print("✅ Database criado")
```

```python
# DDL — criar tabelas com sintaxe SQL nativa

spark.sql("DROP TABLE IF EXISTS local.ecommerce.clientes")
spark.sql("""
    CREATE TABLE local.ecommerce.clientes (
        id_cliente    INT     NOT NULL,
        nome          STRING  NOT NULL,
        email         STRING  NOT NULL,
        cidade        STRING,
        data_cadastro DATE,
        ativo         BOOLEAN
    ) USING iceberg
""")
print("✅ Tabela clientes criada")

spark.sql("DROP TABLE IF EXISTS local.ecommerce.produtos")
spark.sql("""
    CREATE TABLE local.ecommerce.produtos (
        id_produto   INT     NOT NULL,
        nome_produto STRING  NOT NULL,
        categoria    STRING,
        preco        DOUBLE,
        estoque      INT,
        ativo        BOOLEAN
    ) USING iceberg
""")
print("✅ Tabela produtos criada")

spark.sql("DROP TABLE IF EXISTS local.ecommerce.pedidos")
spark.sql("""
    CREATE TABLE local.ecommerce.pedidos (
        id_pedido        INT     NOT NULL,
        id_cliente       INT     NOT NULL,
        id_produto       INT     NOT NULL,
        data_pedido      DATE,
        quantidade       INT,
        valor_total      DOUBLE,
        status           STRING,
        data_atualizacao DATE
    ) USING iceberg
""")
print("✅ Tabela pedidos criada")
```

```python
# UPDATE — Iceberg suporta UPDATE nativo via SQL

spark.sql("""
    UPDATE local.ecommerce.pedidos
    SET status = 'entregue', data_atualizacao = date('2024-02-10')
    WHERE id_pedido = 3
""")
print("✅ UPDATE — pedido 3 atualizado para 'entregue'")
spark.sql("SELECT * FROM local.ecommerce.pedidos").show()

spark.sql("""
    UPDATE local.ecommerce.produtos
    SET preco = 79.90
    WHERE id_produto = 2
""")
print("✅ UPDATE — preço do Mouse atualizado")
spark.sql("SELECT * FROM local.ecommerce.produtos").show()
```

```python
# DELETE

spark.sql("""
    UPDATE local.ecommerce.clientes
    SET ativo = false
    WHERE id_cliente = 4
""")
print("✅ Soft DELETE — cliente 4 desativado")
spark.sql("SELECT * FROM local.ecommerce.clientes").show()

spark.sql("""
    DELETE FROM local.ecommerce.pedidos
    WHERE status = 'cancelado'
""")
print("✅ Hard DELETE — pedidos cancelados removidos")
spark.sql("SELECT * FROM local.ecommerce.pedidos").show()
```

```python
# TIME TRAVEL — por versão (snapshot) e por timestamp

# Ver snapshots disponíveis
print("📜 Histórico de snapshots — tabela pedidos:")
spark.sql("""
    SELECT snapshot_id, committed_at, operation
    FROM local.ecommerce.pedidos.snapshots
""").show(truncate=False)

# Ler versão inicial (snapshot 0)
print("🕐 Pedidos — estado inicial (versão 0):")
spark.sql("""
    SELECT * FROM local.ecommerce.pedidos
    VERSION AS OF 1
""").show()

# Schema evolution — adicionar coluna nova sem reescrever dados
spark.sql("""
    ALTER TABLE local.ecommerce.pedidos
    ADD COLUMN desconto DOUBLE
""")
print("✅ Schema evolution — coluna 'desconto' adicionada sem reescrever dados!")
spark.sql("SELECT * FROM local.ecommerce.pedidos").show()
```
